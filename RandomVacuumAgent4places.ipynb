# Random Vacuum Agent 4 places

Archive [agents.py](agents.py) contains the agent constructor and settings(random). Also, it contains the enviroment base and the Vacuum environment - with four places.
 
This implementation was inspired in this [repository](https://github.com/aimacode/aima-python/blob/master/).

```python
from agents import *
```

#### Creating the Vacuum Environment with four places and checking its status.

```python
loc_A, loc_B, loc_C, loc_D = (0, 0), (1, 0), (0, 1), (1, 1)
trivial_vacuum_env = TrivialVacuumEnvironment()

print("Intial state of the Environment: {}.".format(trivial_vacuum_env.status))
```

#### Creating the agent, that will choose any of the actions from 'Right', 'Left', 'Up', 'Down', 'Suck' and 'NoOp' (No Operation) randomly.

```python
random_agent = Agent(program=RandomAgentProgram(['Right', 'Left', 'Up', 'Down', 'Suck', 'NoOp']))
```

#### Adding the agent to the environment and checking its status.

```python
trivial_vacuum_env.add_thing(random_agent)

print("RandomVacuumAgent is located at {}.".format(random_agent.location))
```

#### Running the environment and checking the current state of the environment.

```python
trivial_vacuum_env.run()

print("State of the Environment: {}.".format(trivial_vacuum_env.status))
print("RandomVacuumAgent is located at {}.".format(random_agent.location))
```
